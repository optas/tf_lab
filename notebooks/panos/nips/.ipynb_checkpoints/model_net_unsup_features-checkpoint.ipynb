```python
def is_centered_in_u_sphere(points, threshold=10e-5):    
    maxima_centered = np.all(np.max(points, 0) + np.min(points, 0) < threshold)    
    in_u_sphere = np.max(l2_norm(points, axis=1)) <= (0.5 + threshold)
    return maxima_centered and in_u_sphere
    
```
