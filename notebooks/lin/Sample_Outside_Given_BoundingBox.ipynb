```python
def readPointFile(filename,Npoint=3000):
    pointlist = np.zeros([Npoint,3],dtype=np.float32)
    numpoint = 0
    for l in open(filename):
        if not l.startswith('#'):
            x,y,z= map(float,l.strip().split())
            pointlist[numpoint][0] = x
            pointlist[numpoint][1] = y
            pointlist[numpoint][2] = z
            numpoint = numpoint + 1
    return pointlist
```
