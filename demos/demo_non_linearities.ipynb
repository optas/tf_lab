```python
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
```

```python
def tanh(x):
    return np.tanh(x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    y = x.copy()
    y[y<0] = 0
    return y

def softplus(x):
    return np.log(1.0 + np.exp(x))
```

```python
x = np.linspace(-5, 5, 100)
funcs = [sigmoid, tanh, relu, softplus]
func_names = ['sigmoid', 'tanh', 'relu', 'soft-plus']

for f in funcs:
    y = f(x)
    plt.plot(x, y, linewidth=3);
    
plt.ylim((-1, 3))
plt.xlim((-5, 5))

plt.xticks = []

plt.legend(func_names, loc='best');
```
